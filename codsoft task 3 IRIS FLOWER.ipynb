```python

data=pd.read_csv("IRIS.csv")
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="sepal_width", data=data)
plt.title("Sepal Width Distribution by Species")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="petal_length", data=data)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
#Scatter Plot with species hue
plt.figure(figsize=(10, 6))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=data, palette="Set1")
plt.title("Sepal Length vs. Petal Length by Species")
plt.show()
```

```python
#Box Plot to visualize the distribution of each feature across different species
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="sepal_length", data=data)
plt.title("Sepal Length Distribution by Species")
plt.show()
```

```python
#Violin Plot to show the distribution and density of data points
plt.figure(figsize=(10, 6))
sns.violinplot(x="species", y="sepal_length", data=data, palette="Set2")
plt.title("Sepal Length Density by Species")
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="petal_width", data=data)
plt.title("Petal Width Distribution by Species")
plt.show()
```

```python


plt.figure(figsize=(10, 6))
sns.violinplot(x="species", y="petal_width", data=data, palette="Set2")
plt.title("Petal Width Density by Species")
plt.show()
```

```python
#Pair Plot to visualize pairwise relationships in the dataset
sns.pairplot(data, hue="species", palette="Set2")

plt.show()
```

```python
#Histogram for each feature
data.hist(figsize=(12, 8), color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions")
plt.show()

```
